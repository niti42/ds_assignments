## Import Necessary libraries:

```python
import pandas as pd
import numpy as np

import re
from nltk.stem import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer

from nltk.corpus import stopwords

from matplotlib import pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
```

## Loading  data and initial analysis.

```python
tweets = pd.read_csv('Elon_musk.csv', names=['tweet_no', 'tweet'])
```

```python
tweets.head()
```

```python
tweets.info()
```

## Data preprocessing
- Remove unwanted columns.
- Remove special characters, numbers etc.
- Lemmatization (stemming if not using lemmatization)
- Get all positive and negative words from lexicon provided.

```python
tweets.drop(index=tweets.index[0], axis=0, inplace=True)
tweets.drop('tweet_no', axis=1, inplace=True)
tweets.reset_index(drop=True, inplace=True)
```

```python
tweets.head()
```

```python
# Function to collect words from lexicons (provided as txt files).
def load_lexicon(filename):
    file = open(filename)

    # Throw away lines of the file until just before the data starts
    # In the provided files, the last line before the actual data starts is a blank line
    while file.readline() != '\n':
        pass

    # Pandas will only process the lines from the current file position onwards
    word_list = np.loadtxt(file, dtype='str')

    # Don't forget to close the file when you're done
    file.close()
    return word_list
    
```

```python
# Get all the positive, negative and stopwords.
positive_words = load_lexicon('positive-words.txt') 
negative_words = load_lexicon('negative-words.txt')
#stopwords = pd.read_csv('stop.txt').to_numpy().ravel() # Get the list of stopwords from the file provided.
```

```python
lemmatizer = WordNetLemmatizer() # lemmatization
#ps = PorterStemmer() # uncomment for stemming

corpus = []
for i in range(0, len(tweets)):
    review = re.sub('RT @[\w]*:', ' ', tweets['tweet'][i]) # Remove twitter return handles (RT @xxx:)
    review = re.sub('@[\w]*', ' ', review) # Remove all twitter handles
    review = re.sub('https?://[A-Za-z0-9./]*', ' ', review) # Remove all URL links
    review = re.sub('[^a-zA-Z]', ' ', review)#remove special characters, numbers, punctuations (except for #)
    review = review.lower()
    review = review.split()
    # Note: I have added the lexicon of stopwords provided inside my nltk lexicon folder since I was getting
    # errors while performing the iterations in the next step when I used the txt file.
    review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english_custom')]
    #review = [ps.stem(word) for word in review if not word in stopwords.words('english_custom')]
    review = ' '.join(review)    
    corpus.append(review)
```

## Visualizing the different words used by Elon Musk to express his sentiments

```python
# Function to visualize the the positive and negative words used by Elon Musk:
def make_word_cloud(corpus, sentiment_words=None, sentiments=True):
    if sentiments == True:
        # Make a list of emotion words and their frequencies per tweet
        word_list = []
        counts = []
        for i in range(len(corpus)): # iterate over each tweet.
            # Search for word from lexicon within each tweet.
            for word in sentiment_words:
                if corpus[i].find(word) != -1: # if the word is found.
                    word_list.append(word)
                    counts.append(corpus[i].count(word)) # no. of times the word occurs

        # Since the wordlist does not take care of one word appearing 
        # twice per sentence, add these words at the end of the list if
        # it repeats.
        for i in range(len(counts)):
            if counts[i] > 1:
                for j in range(counts[i]):
                    word_list.append(word_list[i])

        # Join all the words to make one big string.
        word_list_com = ' '.join(word_list)
    else:
        word_list_com = ' '.join(corpus)
    
    # Make the word cloud.
    wordcloud = WordCloud(max_font_size=50, 
                          max_words=500, 
                          background_color='white', 
                          colormap='tab10', 
                          random_state=42).generate(word_list_com)
    plt.figure()
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
    
    # Note: Other suitable cmaps: 'gist_earth', 'Dark2', 'viridis', 'jet'
```

```python
# Wordcloud for positive words.
make_word_cloud(corpus, positive_words)
```

```python
# Wordcloud for negative words.
make_word_cloud(corpus, negative_words)
```

```python
# Wordcloud for all words used by Elon Musk irrespective of sentiment type.
make_word_cloud(corpus,sentiments=False)
```

## Rule based sentiment analysis

```python
# Create dictionary to hold number of positive and negative words 
# used by Elon Musk per tweet.
sent_col = ["positive","negative"]
sentiment_freq = dict.fromkeys(sent_col)

pos_freq = []
neg_freq = []

for i in range(len(corpus)):
    counts_pw = 0
    counts_nw = 0
    for word in positive_words:
        counts_pw += corpus[i].count(word)
    pos_freq.append(counts_pw)
    
    for word in negative_words:
        counts_nw += corpus[i].count(word)
    neg_freq.append(counts_nw)
    
sentiment_freq["positive"] = pos_freq
sentiment_freq["negative"] = neg_freq

# Add the positive and negative word frequencies per tweet in 
# their respective columns.
tweets_sentiment = pd.DataFrame(sentiment_freq,index=tweets.index)
tweets_df = pd.DataFrame(corpus,columns = ['tweet'],index=tweets.index)
tweets_prc = pd.concat([tweets_df, tweets_sentiment], axis = 1)
```

### Observations:

```python
# Compute the ratio of positive and negative word counts per tweet.
# Ref: https://towardsdatascience.com/design-your-own-sentiment-score-e524308cf787
# Rule: degree of positivity/ negativity = (#of positive words)/ (#of negative words + 1)
tweets_prc1 = tweets_prc.copy()
keys = ['positive_tweets', 'negative_tweets', 'neutral_tweets', 'no_emotion_tweets' ]
emotion_sumry = dict.fromkeys(keys)

# Ratio of positive and negative words count. If ratio is around 1, emotion is neutral
# If ratio > 1, emotion is positive or negative according to metrics
tweets_prc1 = tweets_prc1.assign(ratio_p = lambda x: (x['positive']/(x['negative']+1)))
tweets_prc1 = tweets_prc1.assign(ratio_n = lambda x: (x['negative']/(x['positive']+1)))
```

```python
tweets_prc1 = tweets_prc1.assign(diff = lambda x: (x['ratio_p']-(x['ratio_n'])))
```

```python
# Function to classify sentiment for rule based approach:
def sentiment_classifier(x):
    if x > 0.5:
        return 'positive'
    elif x < -0.5:
        return 'negative'
    else:
        return 'ambiguous'
```

```python
tweets_prc1['sentiment'] = tweets_prc1['diff'].apply(lambda x: sentiment_classifier(x))
```

```python
tweets_prc1.head()
```

### Line chart for visualizing the degree of positivity or negativity values.

```python
def tweet_vis(df, inc, step, labels):
    num_rows = df.shape[0]
    start_ind = inc
    num_pts = start_ind + step
    plt.figure(figsize = (25,4))
    plt.plot(range(num_rows)[start_ind:num_pts], df[labels[0]][start_ind:num_pts], 'g--', label = labels[0], alpha =0.9)
    plt.plot(range(num_rows)[start_ind:num_pts], df[labels[1]][start_ind:num_pts], 'r--', label = labels[1], alpha =0.9)
    plt.plot(range(num_rows)[start_ind:num_pts], df[labels[2]][start_ind:num_pts], 'm-', label = labels[2], alpha = 0.2)

    plt.xlabel('tweet no.')
    plt.ylabel('deg of positivity or negativity')
    plt.legend()
    plt.show()
```

### Bar graph to  summarize of different sentiments expressed.

```python
def sentiment_summary(df):
    summary = dict.fromkeys(df.unique())
    for key in summary.keys():
        summary[key] = df[df==key].count()
    
    summary_df = pd.DataFrame(list(summary.items()),columns=['tweeet_type','no_of_tweets'])
    return summary_df
```

```python
summary_rule_based = sentiment_summary(tweets_prc1['sentiment'])
```

```python
plt.bar(summary_rule_based['tweeet_type'],summary_rule_based['no_of_tweets'])
plt.xlabel('tweet type')
plt.ylabel('no. of tweets')
plt.title('Summary of different sentiments seen in Elon Musk\'s tweets')
plt.show()
```

## Using VADER for sentiment analysis
Ref: https://python.plainenglish.io/twitter-sentiment-analysis-using-vader-tweepy-b2a62fba151e

```python
tweets_prc2 = tweets_prc.copy()
```

```python
tweets_prc2.drop(['positive', 'negative'], axis=1, inplace=True)
```

```python
tweets_prc2.head()
```

```python
# !pip install vaderSentiment
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
scores = []# Declare variables for scores
compound_list = []
positive_list = []
negative_list = []
neutral_list = []
for i in range(tweets_prc2['tweet'].shape[0]):#print(analyser.polarity_scores(sentiments_pd['text'][i]))
    compound = analyzer.polarity_scores(tweets_prc2['tweet'][i])["compound"]
    pos = analyzer.polarity_scores(tweets_prc2['tweet'][i])["pos"]
    neu = analyzer.polarity_scores(tweets_prc2['tweet'][i])["neu"]
    neg = analyzer.polarity_scores(tweets_prc2['tweet'][i])["neg"]
    
    scores.append({"Compound": compound,
                       "Positive": pos,
                       "Negative": neg,
                       "Neutral": neu
                  })
```

```python
# Convert the scores dictionary to dataframe, join it to tweets_prc2 dataframe.
sentiments_score = pd.DataFrame.from_dict(scores)
tweets_prc2 = tweets_prc2.join(sentiments_score)
tweets_prc2.head()
```

```python
# Classify the tweet as positive, negative or neutral based on the compound score.
# Ref: https://akladyous.medium.com/sentiment-analysis-using-vader-c56bcffe6f24
def polarity_score(compound):
    if compound > 0.05:
        return "positive"
    elif compound < -0.05:
        return "negative"
    elif compound >= -0.05 and compound < 0.05:
        return "neutral"
```

```python
tweets_prc2['sentiment'] = tweets_prc2['Compound'].apply(lambda val: polarity_score(val))
```
